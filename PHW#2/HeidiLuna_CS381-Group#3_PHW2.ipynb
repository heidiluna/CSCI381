## Exercise 2:

Explain what's wrong in the following query and provide a correct alternative

```
SELECT Customers.custid, Customers.companyname, Orders.orderid, Orders.orderdate
FROM Sales.Customers AS C
  INNER JOIN Sales.Orders AS O
    ON Customers.custid = Orders.custid;
```

<u>ANSWER:</u> The initial stage of using the JOIN table operator involves assigning the shorter aliases C and O to the Customers and Orders tables. This aliasing simplifies the table names for the duration of the query. During the remaining phases of the logical query processing, the original table names are not available; instead, only the defined aliases can be used.

<u>**CORRECT ALTERNATIVE:**</u>
```
SELECT C.custid, C.companyname, O.orderid, O.orderdate
FROM Sales.Customers AS C
INNER JOIN Sales.Orders AS O
ON C.custid = O.custid;

```

## Exercise 8 (Optional):

Explain why the following query isn't a correct solution query for exercise 7

```
SELECT C.custid, C.companyname, O.orderid, O.orderdate
FROM Sales.Customers AS C
  LEFT OUTER JOIN Sales.Orders AS O
    ON O.custid = C.custid
WHERE O.orderdate = '20160212'
   OR O.orderid IS NULL;

```

**<u>ANSWER:</u>** The outer join connects all customers with their corresponding orders while also including those customers who haven't placed any orders. For customers without orders, the order attributes will show as NULL. The subsequent WHERE clause filters the results to retain only those rows where the order date is February 12, 2022, or where the order ID is NULL (indicating customers without any orders). This filter eliminates customers who made purchases on dates other than February 12, 2022, even if they placed orders on other days. According to Exercise 7, the query should return all customers who ordered on February 12, 2022. The last two rows in the result show customers with no orders at all. It’s important to note that several of the 91 customers in the Customers table are absent from the results because they did not place any orders on February 12, 2022, even if they did order on different dates.

## Proposition 3:

Write a query that retrives information about customers, their orders and order details such as productid and quantity. Customers without any orders will appear in the result set, but the orderid, productid, and qty fields will show as NULL.

````sql
USE TSQLV4
SELECT C.custid, O.orderid, OD.productid, OD.qty
FROM Sales.Customers AS C
  LEFT OUTER JOIN Sales.Orders AS O
    ON C.custid = O.custid
  INNER JOIN Sales.OrderDetails AS OD
    ON O.orderid = OD.orderid;
````

## Proposition 4:

Write a query that returns customers who have not placed any orders. The results should list customers with their customer id and associated company. 

````sql
USE TSQLV4
SELECT C.custid, C.companyname
FROM Sales.Customers AS C
  LEFT OUTER JOIN Sales.Orders AS O
    ON C.custid = O.custid
WHERE O.orderid IS NULL;
````

## Proposition 5:

Write a query that returns a list of customers who have placed an order, along with their order IDs, product IDs, and quantities for each product they have ordered.
If a customer has placed multiple orders or ordered multiple products per order, that customer will appear multiple times in the result set.

````sql
USE TSQLV4
SELECT
  C.custid, C.companyname, O.orderid,
  OD.productid, OD.qty
FROM Sales.Customers AS C
  INNER JOIN Sales.Orders AS O
    ON C.custid = O.custid
  INNER JOIN Sales.OrderDetails AS OD
    ON O.orderid = OD.orderid;
````
